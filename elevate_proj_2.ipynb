```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset (change file name accordingly)
df = pd.read_csv("/content/Housing.csv")
# df = pd.read_csv("medical_appointments.csv")

# Check missing values
missing_values = df.isnull().sum()
print(missing_values)
```

```python
missing_values = missing_values[missing_values > 0]
import matplotlib.pyplot as plt

if not missing_values.empty:
    missing_values.plot(kind='bar')
    plt.title("Missing Values per Column")
    plt.ylabel("Count of Missing Values")
    plt.xlabel("Columns")
    plt.show()
else:
    print("✅ No missing values found — nothing to plot.")
```

```python
# Identify numerical columns
num_cols = df.select_dtypes(include=['int64', 'float64']).columns

# Apply median imputation (preferred for skewed data)
for col in num_cols:
    df[col].fillna(df[col].median(), inplace=True)
```

```python
# Identify categorical columns
cat_cols = df.select_dtypes(include=['object']).columns

# Apply mode imputation
for col in cat_cols:
    df[col].fillna(df[col].mode()[0], inplace=True)
```

```python
threshold = 0.4 * len(df)

cols_to_drop = df.columns[df.isnull().sum() > threshold]
print("Dropped columns:", cols_to_drop)

df.drop(columns=cols_to_drop, inplace=True)
```

```python
# Check remaining missing values
print(df.isnull().sum())

# Dataset info
df.info()
```

```python
df_original = pd.read_csv("/content/Housing.csv")

print("Before Cleaning:")
print("Shape:", df_original.shape)
print("Total Missing Values:", df_original.isnull().sum().sum())

print("\nAfter Cleaning:")
print("Shape:", df.shape)
print("Total Missing Values:", df.isnull().sum().sum())
```
